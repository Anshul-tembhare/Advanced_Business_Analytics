```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
```

```python
data = pd.read_csv('data/crp_cleandata.csv')
```

```python
data.info()
```

```python
pd.tools.plotting.scatter_matrix(data[['Salerank', 'X2013USSales', 'X2013WorldSales', 'ProfitMargin',
                                       'NumStores', 'Reward', 'RewardSize']], diagonal="hist", figsize =(14,14))
plt.show()
```

```python
data.describe()
```

```python
plt.figure(figsize=(9,7))
plt.scatter(x=data['Salerank'], y=data['NumStores'])
plt.xlabel("Reward Size")
plt.ylabel("Profit Margin")
plt.show()
```
